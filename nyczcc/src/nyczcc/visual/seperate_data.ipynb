```python
import numpy as np
import pandas as pd

data = pd.read_csv('/Users/HuanYe/Desktop/Project2_DS504/data/data.csv')
data
```

```python
important_features=['tpep_pickup_datetime','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude']
important_data= data.loc[:,important_features]
important_data
    
```

```python
dt=np.vstack(important_data.tpep_pickup_datetime.astype(str).apply(lambda x:list(map(str,[x[:10],x[11:13]]))).values)
important_data['date']=dt[:,0]
important_data['time']=dt[:,1]
```

```python
important_data=important_data.drop(['tpep_pickup_datetime'],axis=1)
important_data
```
